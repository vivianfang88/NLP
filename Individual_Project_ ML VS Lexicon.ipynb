# Machine Learning

```python
# import data 
import pandas as pd

# hotel_reviews_data is being preprocess by removing stopwords, punctuation and numbers in assignment 2.
hotel_reviews_data = pd.read_csv("assignment4.csv")
hotel_reviews_data.head()
```

```python
# split the data into training and test set 

from sklearn.model_selection import train_test_split

input_data = hotel_reviews_data["Lemmatized"]
label_data = hotel_reviews_data["Sentiment"]

x_train, x_test, y_train, y_test = train_test_split(input_data, label_data, test_size = 0.2, random_state=42)
print (len(x_train))
print (len(x_test))
print (len(y_train))
print (len(y_test))
```

## Bag of words (BOW)

## Doc2vec

```python
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
import nltk
nltk.download('punkt')
from sklearn import utils
```

```python
Doc2vec_data = hotel_reviews_data[["Lemmatized", "Sentiment"]]
train, test = train_test_split(Doc2vec_data, test_size = 0.2, random_state=42)

# print (len(train))
# print (len(test))
```

```python
def tokenize_text(sentence):
    tokens = []
#     print (sentence)
#     for sent in nltk.sent_tokenize(sentence):
#         print (sent)
    for word in nltk.word_tokenize(sentence):
#         print (word)
#         print (len(word))
        if len(word) < 2:
            continue
        tokens.append(word.lower())
#         print (tokens)
    return tokens

# tagged the documents with unique ID
train_tagged = train.apply(lambda x : TaggedDocument(words = tokenize_text(x["Lemmatized"]),tags=[x.Sentiment]), axis=1)
print (train_tagged)
test_tagged = test.apply(lambda x : TaggedDocument(words = tokenize_text(x["Lemmatized"]),tags=[x.Sentiment]), axis=1)
```

```python
# build vocab

model_dbow = Doc2Vec(dm=0, vector_size=300, negative =5, hs=0, min_count =1, sample=0, workers = 8, alpha=0.025, min_alpha=0.001)
model_dbow.build_vocab([x for x in train_tagged])

# train model

for epoch in range (50):
    model_dbow.train(utils.shuffle([x for x in train_tagged]), total_examples = len(train_tagged.values), epochs =1)
    model_dbow.alpha -= 0.002
    model_dbow.min_alpha = model_dbow.alpha
    
# create final vector feature

def vec_for_learning(model, tagged_docs):
    sents = tagged_docs
    targets, regressors = zip(*[(doc.tags[0], model.infer_vector(doc.words, steps=20)) for doc in sents])
#     print (targets, regressors)
    return targets, regressors

Doc2vec_y_train, Doc2vec_x_train = vec_for_learning(model_dbow, train_tagged)
# print (Doc2Vec_x_train)
Doc2vec_y_test, Doc2vec_x_test = vec_for_learning(model_dbow, test_tagged)
```

### Classification_Doc2vec

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB

NB_Doc2vec = GaussianNB()
NB_Doc2vec.fit(Doc2vec_x_train, Doc2vec_y_train)
predict_NB_Doc2vec = NB_Doc2vec.predict(Doc2vec_x_test)
result_NB_Doc2vec = classification_report(predict_NB_Doc2vec, Doc2vec_y_test)
print("NB Classifier result:")
print (result_NB_Doc2vec)

# SVM
SVM_Doc2vec = svm.SVC(gamma= "scale")
SVM_Doc2vec.fit(Doc2vec_x_train, Doc2vec_y_train)
predict_SVM_Doc2vec = SVM_Doc2vec.predict(Doc2vec_x_test)
result_SVM_Doc2vec = classification_report(predict_SVM_Doc2vec, Doc2vec_y_test)
print("SVM Classifier result:")
print (result_SVM_Doc2vec)

# Logistic Regression
LR_Doc2vec = svm.SVC(gamma= "scale")
LR_Doc2vec.fit(Doc2vec_x_train, Doc2vec_y_train)
predict_LR_Doc2vec = LR_Doc2vec.predict(Doc2vec_x_test)
result_LR_Doc2vec = classification_report(predict_LR_Doc2vec,  Doc2vec_y_test)
print("LR Classifier result:")
print (result_LR_Doc2vec)
```

```python
import pandas as pd
```

```python
# import +ve dataset

mydata1 = pd.read_csv("Hotel +ve reviews.csv")
mydata1.head()
```

```python
# import -ve dataset

mydata2 = pd.read_csv("Hotel -ve reviews.csv")
mydata2.head()
```

```python
# merge +ve&-ve files 

mydata = mydata1.append(mydata2).reset_index(drop=True)
pd.set_option("display.max_rows", None)
mydata
```

```python
newdata = mydata.drop(["Malay Comments", "Bing Translate"], axis=1)
newdata.head()
```

```python
#change column name
Column_name = ['Google_Translate']
newdata.columns = Column_name
newdata.head()
```

```python
# create checkpoint before preprocess 

newdata2 = newdata
```

```python
# preprocess functions to remove stopwords & punctuation & numbers

import string 
from nltk.corpus import stopwords
import re

new_stopwords = set(stopwords.words("english")) - set(["not"])
# print(new_stopwords)

def text_process(x):
    
    #no_punc = []    
    nopunc = [char for char in x if char not in string.punctuation]
    nopunc = "".join(nopunc)
    
    no_punc = [char for char in nopunc.split() if char.lower() not in new_stopwords]
    return no_punc
    
    


def remove_numbers(x):
    
    no_punc = text_process(x)
    reviews = reviews = re.sub("[^a-zA-Z]", " ", str(no_punc))
    return reviews    
   
```

```python
# preprocess 

clean_text = newdata2["Google_Translate"].apply(remove_numbers)
newdata2["Clean_Text"] =pd.DataFrame(clean_text)
newdata2.head()
```

```python
#create another checkpoint 

newdata3 = newdata2
newdata3.head()
```

```python
# insert sentiment labels for accuracy test later on [label comparison] 

newdata3.loc[newdata.index[0:51], "Sentiment"] = ["Positive"]
newdata3.loc[newdata.index[51:], "Sentiment"] = ["Negative"]
newdata3
```

```python
#import lexicon file 

lexicon_sentiment = pd.read_csv("lexicon_sentiment_file_edited.csv")
lexicon_sentiment.head()
```

```python
#Calculate sum score & sentiment 

def sentiment(sentence, score):
    print ("Sentence" + sentence)
    print ("Score:" + str(score))
    if (score > 0):
        print('Review sentiment: POSITIVE')
        senti = 'Positive'
    elif (score < 0):
        print('Review sentiment: NEGATIVE')
        senti = 'Negative'
    else:
        print('Review sentiment: NEUTRAL')
        senti = 'Neutral'
    print ("")
    return senti



def whole_reviews(review):
    sentence_array = []
    score_array = []
    sentiment_array = []
    
    for sentences in newdata3.Clean_Text:
        score = 0

        for word in sentences.split():

            if word in lexicon_sentiment["ADJ_Text"].values:
                score += lexicon_sentiment.query("ADJ_Text == @word")["Score"].values[0]


        sentence_array.append(sentences)
        score_array.append(score)
        senti = sentiment(sentences, score)
        sentiment_array.append(senti)
    return sentence_array, score_array, sentiment_array
    
    

sentence, score, senti = whole_reviews(newdata3["Clean_Text"])
```

```python
# create a data frame for accuracy test later on.

lexicon_result = pd.DataFrame()
lexicon_result['Sentence'] = pd.Series(sentence).values
lexicon_result['Score'] = pd.Series(score).values
lexicon_result['Sentiment'] = pd.Series(senti).values

lexicon_result
```

```python
# accuracy test 
import numpy as np

#add the sentiment column from newdata3 dataframe into lexicon_result dataframe
lexicon_result['Original_Sentiment'] = newdata3['Sentiment'] 
# lexicon_result.head()

lexicon_result["Sentiment_Match"] = np.where(lexicon_result["Sentiment"] == newdata3["Sentiment"], "True", "False")
lexicon_result
```

```python
# get the freq of ["True/False"] values based on original sentiment

result = lexicon_result.groupby('Original_Sentiment').Sentiment_Match.value_counts()
result
```

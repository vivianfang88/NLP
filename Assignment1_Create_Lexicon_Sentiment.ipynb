```python
# #preprocess
# #remove punctuation and stopwords.
# #reduce computational cost

import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

new_english = set(stopwords.words('english')) - set(["not"]) # restricting the "not" words being removed
# print (new_english)
# print (set(stopwords.words('english')))

def text_process(x):
    
    nopunc = [char for char in x if char not in string.punctuation]
    nopunc = "".join(nopunc)
    nopunc_list = [words for words in nopunc.split() if words.lower() not in new_english]

    return nopunc_list
    


    

    
nopunc =pos_data["Google_Translate"].apply(text_process)   
nopunc
```

```python
# convert the data to list type

pos_data3 = pos_data2["Clean_Text"].values.tolist()
pos_data3
```

```python
import spacy 
nlp = spacy.load("en_core_web_sm")

clean_pos_text = []
for list in pos_data3:
    for strings in list:
        doc = nlp(strings)
        for oktoken in doc:
#             print(oktoken.text, oktoken.pos_)
            if oktoken.pos_ == "ADJ":
                clean_pos_text.append((oktoken.text, 1))
            if oktoken.pos_ == "PART":
                if oktoken.text == 'not':
                    clean_pos_text.append((oktoken.text, 1))
```

```python
# convert the list to df for later CSV file conversion
import numpy as np

hotel_pos= pd.DataFrame(np.array(clean_pos_text), columns = ["ADJ_Text", "Score"])
# pd.set_option("display.max_rows", None)
hotel_pos
```

```python
# remove the repeated ADJ

hotel_pos = hotel_pos.drop_duplicates(subset= "ADJ_Text", keep= "first")
hotel_pos2 = hotel_pos.reset_index(drop=True)
hotel_pos2
```

```python
## proceed with negative hotel reviews

neg_data= pd.read_csv('Hotel -ve reviews.csv')
neg_data.head()
```

```python
neg_data = neg_data.drop(['Malay Comments', 'Bing Translate'], axis=1)
Column_names = ['Google_Translate']
neg_data.columns = Column_names
neg_data.head()
```

```python
no_punc_neg = neg_data["Google_Translate"].apply(text_process)
no_punc_neg[0:5]
# neg_data check manually
# neg_data["check"] = pd.DataFrame(check2)
# neg_data.head()
```

```python
#create checkpoint 

neg_data2 = no_punc_neg
neg_data2 = neg_data2.values.tolist()
neg_data2[0:2]
```

```python


clean_neg_text = []

for list in neg_data2:
    for strings in list:
        doc = nlp(strings)
        for oktoken in doc:
            print(oktoken.text, oktoken.pos_)
            if oktoken.pos_ == "ADJ":
                clean_neg_text.append((oktoken.text, -1))
            if oktoken.pos_ == "PART": # extract the word "not" ; strong negative sentiment
                if oktoken.text == 'not':
                    clean_neg_text.append((oktoken.text, -1))
```

```python
hotel_neg= pd.DataFrame(np.array(clean_neg_text), columns = ["ADJ_Text", "Score"])
# pd.set_option("display.max_rows", None)
hotel_neg
```

```python
# remove duplicates

hotel_neg = hotel_neg.drop_duplicates(subset= "ADJ_Text", keep= "first")
hotel_neg2 = hotel_neg.reset_index(drop = True)
hotel_neg2.head()
```

```python
# combine both pos&neg files

combine_files = hotel_pos2.append(hotel_neg2).reset_index(drop=True)
# print(combine_files)


lexicon_sentiment = pd.DataFrame(np.array(combine_files), columns = ["ADJ_Text", "Score"])
lexicon_sentiment.head()


```
